```python
from transformers import AutoTokenizer, AutoModel, utils
utils.logging.set_verbosity_error()  # Suppress standard warnings
tokenizer1 = AutoTokenizer.from_pretrained("models/model1",padding=True,  max_length=512)
model1 = AutoModel.from_pretrained("models/model1", output_attentions=True)
txt = "The cat sat on the mat"
out = model1(tokenizer1.encode(txt, return_tensors='pt'))
```

```python
decoded_text = tokenizer1.decode(tokenizer1.encode(txt, return_tensors='pt')[0])
print(decoded_text)
```

```python
model1(tokenizer1.encode(txt, return_tensors='pt'))[-1][0][0][0][0][128:132]
```
